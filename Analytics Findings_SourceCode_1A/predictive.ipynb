```python
import networkx as nx
import csv
import json
import matplotlib.pyplot as plt
import itertools
```

```python
with open("edge_head_to_head.csv", "r", encoding="utf-8") as file:
    file = file.readlines()
    matches = [[x.split(",")[1].strip(), x.split(",")[3].strip()] for x in file[1:]]
```

```python
matches
```

```python
# Fetch market values datas
with open("vertex_head_to_head.csv", "r", encoding="utf-8") as file:
    file = file.readlines()
    clubs = [x.split(",")[1].strip() for x in file[1:]]
```

```python
clubs
```

```python
# Creating Graph

G = nx.Graph()

G.add_nodes_from(clubs)
G.add_edges_from(matches)
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
#Community Detection with Girvan Newman

from networkx.algorithms import community
import matplotlib.pyplot as plt

comm_det_gir = community.girvan_newman(G)
```

```python
#Print Community
for communities in itertools.islice(comm_det_gir, 5):
    print(tuple(sorted(c) for c in communities))  
```

```python
top_level_comm_bem = next(comm_det_gir)
next_level_comm_bem = next(comm_det_gir)

print("Community Detection in League")
print("Count:", len(sorted(map(sorted, next_level_comm_bem))))
sorted(map(sorted, next_level_comm_bem))
```
